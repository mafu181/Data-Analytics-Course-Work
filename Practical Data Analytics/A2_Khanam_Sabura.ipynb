### Question 4:

```python
cat_cols = [column for column in df.columns if df[column].dtypes == object]

df = pd.get_dummies(df, columns=cat_cols)

df
```

### Question 5:

```python
df_x = df.drop('age', axis=1)
df_y = df['age']
```

```python
from sklearn.model_selection import train_test_split

trainX, testX, trainY, testY = train_test_split(df_x, df_y, test_size=0.3, random_state = 0)

trainX.shape, testX.shape, trainY.shape, testY.shape
```

```python
trainX_prc = trainX.copy()
trainX_prc
```

```python
testX_prc = testX.copy()
testX_prc
```

### Question 6:

```python
max_avg_glucose_level_train = max(trainX['avg_glucose_level'])
#max_avg_glucose_level_train
max_avg_glucose_level_test = max(testX['avg_glucose_level'])
#max_avg_glucose_level_test
max_avg_glucose_level = max(max_avg_glucose_level_train, max_avg_glucose_level_test)
max_avg_glucose_level
```

```python
trainX_prc["avg_glucose_level"] = trainX_prc["avg_glucose_level"] / max_avg_glucose_level
trainX_prc
```

```python
testX_prc["avg_glucose_level"] = testX_prc["avg_glucose_level"] / max_avg_glucose_level
testX_prc
```

### Question 7:

```python
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix
tree = DecisionTreeClassifier(random_state=0)
tree.fit(trainX_prc, trainY)
tree.score(trainX_prc, trainY) 
```

```python
x = f1_score(trainY, tree.predict(trainX_prc))
v = f1_score(testY, tree.predict(testX_prc))
u = accuracy_score(trainY, tree.predict(trainX_prc))
w = accuracy_score(testY, tree.predict(testX_prc))

print("Train f1 score:", x)
print("Train accuracy score:", u)
print("Test f1 score:", v)
print("Test accuracy score:", w)
```

### Question 8:

```python
for md in range(1,26):
  model = DecisionTreeClassifier(max_depth=md, random_state=0) 
  model.fit(trainX_prc, trainY)
  print('Training Accuracy for max_depth {} is: '.format(md), model.score(trainX_prc, trainY))
  print('Test Accuracy for max_depth {} is:'.format(md), model.score(testX_prc,testY))
  print ('')
 
```

### Question 9:

```python
best_depth = 7
for i in [100,400,1000,3000, len(trainX)]:
    trainX_i = trainX.iloc[1:i+1]
    trainY_i = trainY.iloc[1:i+1]
    model = DecisionTreeClassifier(max_depth= best_depth, random_state=42) 
    model.fit(trainX_i, trainY_i)
    print(f'Training Accuracy for sample size {i} is: {model.score(trainX_i, trainY_i)}')
    print('Test Accuracy for sample size {} is:'.format(i), model.score(testX_prc,testY))
    print('')
```
