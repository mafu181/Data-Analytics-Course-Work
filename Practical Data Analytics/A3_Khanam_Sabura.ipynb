# Question 3:

```python
x = df.drop(['MEDV'], axis=1)
y = df['MEDV']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import ShuffleSplit
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.25, random_state=0)
train_x.shape, test_x.shape, train_y.shape, test_y.shape
```

# Question 4:

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np

X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  
y = np.array([2, 8, 18, 32, 50])  

pipeline = Pipeline([
    ("poly_features", PolynomialFeatures(degree=2, include_bias=False)),
    ("scaler", StandardScaler()),
    ("linear_regression", LinearRegression())
])
```

# Question 5:

```python
pipeline.fit(train_x, train_y)
predictions = pipeline.predict(test_x)

mse = mean_squared_error(test_y, predictions)
print(f"Mean Squared Error on Test Data: {mse}")
```

# Question 6:

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso, Ridge, ElasticNet
 
lasso = Lasso()
ridge = Ridge()
elastic_net = ElasticNet()

param_grid_lasso = {'alpha': np.linspace(0.1, 2, 20)} 
param_grid_ridge = {'alpha': np.linspace(0.1, 2, 20)}  
param_grid_elastic_net = {
    'alpha': np.linspace(0.1, 2, 20),  
    'l1_ratio': np.linspace(0.1, 0.9, 9)  
}
grid_search_lasso = GridSearchCV(lasso, param_grid_lasso, cv=5, scoring='neg_mean_squared_error')
grid_search_ridge = GridSearchCV(ridge, param_grid_ridge, cv=5, scoring='neg_mean_squared_error')
grid_search_elastic_net = GridSearchCV(elastic_net, param_grid_elastic_net, cv=5, scoring='neg_mean_squared_error')

grid_search_lasso.fit(train_x, train_y)
grid_search_ridge.fit(train_x, train_y)
grid_search_elastic_net.fit(train_x, train_y)

best_lasso = grid_search_lasso.best_estimator_
best_ridge = grid_search_ridge.best_estimator_
best_elastic_net = grid_search_elastic_net.best_estimator_

lasso_predictions = best_lasso.predict(test_x)
ridge_predictions = best_ridge.predict(test_x)
elastic_net_predictions = best_elastic_net.predict(test_x)

lasso_mse = mean_squared_error(test_y, lasso_predictions)
ridge_mse = mean_squared_error(test_y, ridge_predictions)
elastic_net_mse = mean_squared_error(test_y, elastic_net_predictions)
```

# Question no 7:

```python
print(f"Lasso MSE on Test Data: {lasso_mse}")
print(f"Ridge MSE on Test Data: {ridge_mse}")
print(f"Elastic Net MSE on Test Data: {elastic_net_mse}")
```
